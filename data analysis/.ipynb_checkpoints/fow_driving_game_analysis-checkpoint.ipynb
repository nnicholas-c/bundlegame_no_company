```python
def pick_first_job(row):
    jobpick_pattern = r'"event":"i:\(\d+,\d+\)\sJob\s(.*?)\snow\savailable".*?"event":"Task\s*(.*?)\s*was started'
    matches = re.findall(jobpick_pattern, row)
    count = 0
    total = len(matches)
    for (a, b) in matches:
        if a == b:
            count += 1
    return count/total
```

```python
complete['Choose First Job'] = complete['GameResult'].apply(pick_first_job)
complete
```

```python
# if didn't pick first job (regex doesn't capture all)
def notpick_first_job(row):
    wait_time = []
    pattern = r'{"time":.*?,"event":"Entered\shome\sscreen,\sdisplayed\sjobs:.*?,"event":"Rated\sthemselves:\s\d+"}'
    matches = re.findall(pattern, row)
    for match in matches:
        jobpick_pattern = r'{"time":(\d+\.\d+),"event":"i:\(\d+,\d+\)\sJob\s(.*?)\snow\savailable".*?,{"time":(\d+\.\d+),"event":"Task\s*(.*?)\s*was started'
        jobs = re.findall(jobpick_pattern, match)
        if len(jobs) > 0 and len(jobs[0]) > 3:
            if jobs[0][1] != jobs[0][3]:
                # waiting time
                wait_time.append(float(jobs[0][2]) - float(jobs[0][0]))
    print(wait_time)
    if len(wait_time) > 0:
        return np.mean(wait_time)
```
